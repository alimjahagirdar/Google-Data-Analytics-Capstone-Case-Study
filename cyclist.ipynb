```python
#reading 12 csv files
import pandas as pd
df1 = pd.read_csv("C:/Users/alimj/Desktop/case study 1/12months/202201-divvy-tripdata.csv")
df2= pd.read_csv("C:/Users/alimj/Desktop/case study 1/12months/202202-divvy-tripdata.csv")
df3= pd.read_csv("C:/Users/alimj/Desktop/case study 1/12months/202203-divvy-tripdata.csv")
df4= pd.read_csv("C:/Users/alimj/Desktop/case study 1/12months/202204-divvy-tripdata.csv") 
df5= pd.read_csv("C:/Users/alimj/Desktop/case study 1/12months/202205-divvy-tripdata.csv")
df6= pd.read_csv("C:/Users/alimj/Desktop/case study 1/12months/202206-divvy-tripdata.csv")
df7= pd.read_csv("C:/Users/alimj/Desktop/case study 1/12months/202207-divvy-tripdata.csv")
df8= pd.read_csv("C:/Users/alimj/Desktop/case study 1/12months/202208-divvy-tripdata.csv")
df9= pd.read_csv("C:/Users/alimj/Desktop/case study 1/12months/202209-divvy-publictripdata.csv")
df10= pd.read_csv("C:/Users/alimj/Desktop/case study 1/12months/202210-divvy-tripdata.csv")
df11= pd.read_csv("C:/Users/alimj/Desktop/case study 1/12months/202211-divvy-tripdata.csv")
df12= pd.read_csv("C:/Users/alimj/Desktop/case study 1/12months/202212-divvy-tripdata.csv")
```

```python
#combinning all files in one dataframe
df = pd.concat([df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11,df12])
```

```python
# changing the datatypes as datetime
df['started_at']= pd.to_datetime(df['started_at'])
df['ended_at']= pd.to_datetime(df['ended_at'])
```

```python
df.info()
```

```python
# extracting the year, month, and day from column
df['year'] = df['started_at'].dt.year
df['month'] = df['started_at'].dt.month_name()
df['week_day'] = df['started_at'].dt.day_name()
df = df.astype({'year':'int16'})
```

```python
df = df.drop(['start_lat', 'start_lng','start_station_id','end_station_id','end_lat','end_lng'], axis=1)
```

```python
df = df.reset_index(drop=True)
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
```
